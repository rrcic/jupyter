```python
from sklearn import tree
import pandas as pd


def createDataSet():
    dataSet = [[0, 0, 0, 0, 'no'],
               [0, 0, 0, 1, 'no'],
               [0, 1, 0, 1, 'yes'],
               [0, 1, 1, 0, 'yes'],
               [0, 0, 0, 0, 'no'],
               [1, 0, 0, 0, 'no'],
               [1, 0, 0, 1, 'no'],
               [1, 1, 1, 1, 'yes'],
               [1, 0, 1, 2, 'yes'],
               [1, 0, 1, 2, 'yes'],
               [2, 0, 1, 2, 'yes'],
               [2, 0, 1, 1, 'yes'],
               [2, 1, 0, 1, 'yes'],
               [2, 1, 0, 2, 'yes'],
               [2, 0, 0, 0, 'no']]
    labels = ['age', 'haveWork', 'haveHouse', 'credit']
    return dataSet, labels


if __name__ == '__main__':
    dataSet, Labels = createDataSet()
    print(dataSet)
    data_target = []  
    for each in dataSet:
        data_target.append(each[-1])
    print(data_target)

    data_list = []  
    data_dict = {}  
    for each_label in Labels:  
        for each in dataSet:
            data_list.append(each[Labels.index(each_label)])
        data_dict[each_label] = data_list
        data_list = []
    print(data_dict)  
    data_pd = pd.DataFrame(data_dict)
    print(data_pd)  

    clf = tree.DecisionTreeClassifier()  
    clf = clf.fit(data_pd.values.tolist(), data_target) 
    
    tree.plot_tree(clf,feature_names = Labels);

    print(clf.predict([[1, 1, 1, 0]]))
```
